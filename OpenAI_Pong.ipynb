```
import gym
import numpy as np
#Neural network
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras import Input
from tensorflow.keras.optimizers import Adam
#Agent
from rl.agents import DQNAgent
from rl.policy import BoltzmannQPolicy
from rl.memory import SequentialMemory
```

```
def build_model(states, actions):
  model = Sequential()
  model.add(Flatten(input_shape=(4,states)))
  model.add(Dense(24, activation="relu"))
  model.add(Dense(24, activation="relu"))
  model.add(Dense(actions, activation="linear"))
  return model
```

```
def build_agent(model, actions):
  policy = BoltzmannQPolicy()
  memory = SequentialMemory(limit=5000, window_length=4)
  dqna = DQNAgent(model=model, memory=memory, policy=policy, nb_actions=actions, nb_steps_warmup=10, target_model_update=1e-2)
  return dqna
```

```
env = gym.make("Pong-ram-v0")
states = env.observation_space.shape[0]
actions = env.action_space.n

model = build_model(states, actions)
dqna = build_agent(model, actions)
dqna.compile(Adam(lr=1e-3), metrics=["mae"])
dqna.fit(env, nb_steps=50000, visualize=False, verbose=1)
```

```
scores = dqna.test(env, nb_episodes=100, visualize=False)
print(np.mean(scores.history['episode_reward']))
```
